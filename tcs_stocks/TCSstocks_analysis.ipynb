```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
data = pd.read_csv("TCS_stock_history.csv")
print(data)
```

```python
data.head()
```

```python
sns.set(rc={'figure.figsize' :(10,5)})
```

```python
data['Data'] = pd.to_datetime(data['Date'])
df = data.set_index('Date')
df.head()
```

```python
sns.lineplot(x=df.index, y=df['Volume'], label='Volume')
plt.show()
```

```python
df.plot(y=['High','Close','Open'], title = 'TCS STOCK PRICE')
```

```python
highest_stockprice = df.sort_values(by = ["High"], ascending = False)
highest = highest_stockprice.head(5)
highest['High']
```

```python
lowest_stockprice = df.sort_values(by = ["Low"], ascending = True)
lowest = lowest_stockprice.head(5)
lowest['Low']
```
